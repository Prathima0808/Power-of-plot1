## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse


# Merge this group df with the original dataframe to get the tumour volume at the last timepoint

#groupby("Mouse ID")["timepoint"].max()
greatest_tp = cleaned_df.groupby("Mouse ID")["Timepoint"].max()

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
merged_df = cleaned_df.merge(greatest_tp,on=["Mouse ID","Timepoint"],how="inner")
merged_df
```

```python
#displayed 50 records.
#merged_df.tail(50)
```

```python
# Put treatments into a list for for loop (and later for plot labels)


# Create empty list to fill with tumour vol data (for plotting)


# Calculate the IQR and quantitatively determine if there are any potential outliers. 

    
    # Locate the rows which contain mice on each drug and get the tumour volumes
    
    
    # add subset 
    
    
    # Determine outliers using upper and lower bounds
    
    # Put treatments into a list for for loop (and later for plot labels)
treatments=["Capomulin" , "Ramicane" , "Infubinol" , "Ceftamin"]
tumor_vol=[]
    
for name in treatments:
        
    final_volume = merged_df.loc[merged_df["Drug Regimen"]== name, "Tumour Volume (mm3)"]
    tumor_vol.append(final_volume)
    
    quartiles = final_volume.quantile([.25 ,.5,.75])
    
    # Determine outliers using upper and lower bounds
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    
    iqr = upperq - lowerq
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    #print the outlines
    outliers = final_volume.loc[(final_volume < lower_bound) | (final_volume > upper_bound)]
    print(f"The {name}'s potential outliers: {outliers}")
    
    
    #reference: https://matplotlib.org/gallery/pyplots/boxplot_demo_pyplot.html#sphx-glr-gallery-pyplots-boxplot-demo-pyplot-py
    
```

```python
# Generate a box plot of the final tumour volume of each mouse across four regimens of interest

plt.boxplot(tumor_vol,labels=treatments,flierprops={'markerfacecolor':'blue','markersize':8})
plt.title ('Final Tumor Volume of each Mouse')
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumour volume for the Capomulin regimen
```

```python
corr_coef = round(st.pearsonr(wt, vol)[0],2)
print(f"The Average Tumor Volume and Mouse Weight is {corr_coef}")
```

```python
#Create a linear regression model
slope, intercept, rvalue, pvalue, std_err = st.linregress(wt, vol)

#Create the equation for the line
line = slope * wt + intercept
line_eq = f"y = {round(slope,2)}x + {round(intercept,2)}" 


#Plot the linear model and generating the size of X and Y axis

fig, ax = plt.subplots(figsize=(7,7))
ax.set_autoscale_on(True)
ax.scatter(wt, vol, marker="o", facecolors="grey", edgecolors="black")
ax.legend(["Mouse"],loc="upper left", title="Regimen: Capomulin")
plt.plot(wt, line, 'b-')

#title of X and Y axis
plt.title("Avg Tumor Volume (mm3) VS Mouse Weight (g)")
plt.ylabel("Avg Tumor Volume (mm3)")
plt.xlabel("Mouse Weight (g)")
plt.ylim(vol.min() - 1, vol.max() + 1)
plt.xlim(wt.min() - 1, wt.max() + 1)
plt.annotate(line_eq, (20,35), fontsize=14, color="black")

plt.show()
```
